## Criando a coluna Pontos_extras

```python
df.head()
```

```python
df['Pontos_extras'] = df['Notas'].apply(lambda x: x*0.4)
df.head()
```

## Criando a coluna Notas_finais

```python
df['Notas_finais'] = df['Notas'] + df['Pontos_extras']
df.head()
```

##Criando a coluna Aprovado_final

```python
df['Aprovado_final'] = df['Notas_finais'].apply(lambda x: True if x > 6 else False)
df.head()
```

## Verificar que não tinha sido aprovado e agora foi após a nota final

```python
df.query('Aprovado == False & Aprovado_final == True')
```
